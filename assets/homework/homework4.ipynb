4. (20 points) Cell-centered finite difference solution of the steady-state 1D Richards equation (4 boxes).

   Let $z$ be positive upward. Consider the steady-state pressure head-based 1D Richards equation

   $$
   \frac{d}{dz}\left[K(h)\frac{dh}{dz}\right]+\frac{dK(h)}{dz}=0.
   $$

   on the domain $z\in[0,1.0]$ m with boundary conditions $h(0)=0$ m and $h(1.0)=-1.0$ m.

   Use the following sand hydraulic properties with the van Genuchten-Mualem model:

   $$
   \theta_r = 0.045,\qquad
   \theta_s = 0.43,\qquad
   \alpha = 0.145\ \text{cm}^{-1},\qquad
   n = 2.68,\qquad
   m = 1-\frac{1}{n}=0.627,\qquad
   l = 0.5,\qquad
   $$

   $$
   K_s = 712.8\ \text{cm/day} = 29.7\ \text{cm/h} = 8.25\times10^{-5}\ \text{m/s}.
   $$

   For the nonlinear system in part (b), use

   $$
   \theta(h)=\theta_r+\frac{\theta_s-\theta_r}{\left(1+|\alpha h|^n\right)^m}, \qquad h<0,
   $$

   with $\theta=\theta_s$ for $h\ge 0$, and

   $$
   K(S_e)=K_s\,S_e^l\left[1-\left(1-S_e^{1/m}\right)^m\right]^2,
   \qquad
   S_e=\frac{\theta-\theta_r}{\theta_s-\theta_r}.
   $$

   Discretize the domain into 4 uniform boxes ($\Delta z=0.25$ m) using a cell-centered finite difference scheme, with unknown pressure heads $h_1,h_2,h_3,h_4$ at the cell centers. For this problem, assume $h_1=h(0.125)=-0.125$ m and $h_1=h(0.875)=-0.875$ m to represent the boundary condition, as an approximation. 

   Evaluate the conductivity at each interior face by arithmetic averaging of the conductivities at the two adjacent cell centers. At the boundary faces, evaluate conductivity directly from the boundary heads $h=0$ m and $h=-1.0$ m.

   For the frozen-conductivity calculation in part (c), assume a linear initial head profile so that

   $$
   h_1=-0.125\ \text{m},\qquad h_2=-0.375\ \text{m},\qquad h_3=-0.625\ \text{m},\qquad h_4=-0.875\ \text{m}.
   $$

   This gives the following frozen face conductivities:

   $$
