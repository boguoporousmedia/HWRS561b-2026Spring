**Question 1. Taylor-Aris dispersion**

Explain the insights from Taylor-Aris dispersion. Use these insights to argue why we may describe solute transport in a porous medium using an advection-dispersion equation.

**Question 2. Richards' equation**

- **a.** Consider a special two-phase system: air and water. Explain the assumptions and line of reasoning used to derive Richards' equation from the two-phase flow equations.

- **b.** If, for some reason, the density of air becomes comparable to that of water, while everything else remains the same, is Richards' equation still valid for describing water flow in the vadose zone? If not, how would you modify Richards' equation (or related terms) for this air-water system?

**Question 3. Extended Darcy's law**

Describe the extended Darcy's law for unsaturated flow, and then write down Darcy's law for a free-drainage condition.

**Question 4. Relative permeability in two-phase flow**

Explain how and why the relative permeability functions differ between wetting and nonwetting phases.

**Question 5. Soil-water characteristic curves (SWCCs)**

- **a.** Draw representative SWCCs for a sand soil and a clay soil, and explain the key differences between the two curves.

- **b.** Explain why the SWCC for a soil has the same shape as the water saturation profile above the water table under hydrostatic conditions.

**Question 6. Governing equation for two-phase flow**

Conceptually describe the steps used to derive the governing equation for two-phase flow. Identify the key conceptual difference between two-phase flow equations and single-phase (saturated) flow equations.

**Question 7. Breakthrough curves under a step-change inlet boundary condition**

For solute transport in a 1D saturated column under steady-state flow with a step-change inlet boundary condition:

- **a.** Draw the expected breakthrough curves for conservative solute transport and for linear, Freundlich, and Langmuir sorption isotherms. Explain the key differences between the breakthrough curves.

- **b.** Draw the expected breakthrough curves for equilibrium and kinetic linear sorption isotherms. Explain the key differences.
